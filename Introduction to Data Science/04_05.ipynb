```python
# import pandas library 

import pandas as pd
```

```python
# load the data from the csv file into a dataframe

us_babies = pd.read_csv("us_baby_names.csv")
```

```python
# display the dataframe

us_babies
```

```python
# slice out the rows containing your name
# save the resulting dataframe in a new variable

us_man = us_babies.loc[us_babies["Name"] == "Manoj", :]
```

```python
# display the new dataframe

us_man
```

```python
# create a horizontal bar plot that displays the counts of your name over the years

us_man.plot.barh(x = "Year", y = "Count")
```
