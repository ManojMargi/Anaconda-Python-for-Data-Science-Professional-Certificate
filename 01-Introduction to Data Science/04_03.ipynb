```python
# import pandas library 

import pandas as pd
```

```python
# load the data from the csv file into a dataframe

states_babies = pd.read_csv("state_baby_names.csv")
```

```python
# display the dataframe

states_babies.head()
```

```python
# for all the rows where the year is 2014, access all the columns
# save the resulting dataframe in a new variable

states_babies_2014 = states_babies.loc[states_babies["Year"] == 2014, :]
```

```python
# display the new dataframe

states_babies_2014
```

```python
# for all the rows of the states_babies_2014 dataframe where the state is CA, access all the columns
# save the resulting dataframe in a new variable

ca_babies_2014 = states_babies_2014.loc[states_babies_2014["State"] == "CA", :]
```

```python
# display the new dataframe

ca_babies_2014
```

```python
# sort the rows of the ca_babies_2014 dataframe in descending order by the "Count" column
# save the sorted dataframe in a new variable

sorted_ca_2014 = ca_babies_2014.sort_values("Count", ascending = False)
```

```python
# display the sorted dataframe

sorted_ca_2014
```

```python
# access the first five rows of the sorted_ca_2014 dataframe

sorted_ca_2014.iloc[0:5]
```
